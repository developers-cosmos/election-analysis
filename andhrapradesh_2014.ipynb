```python
def sortdict(given_dict):
    final = sorted(given_dict.items(), key = lambda kv:(kv[1], kv[0]), reverse = True)
    return final[0][0]
```

```python
data = pd.read_csv("andhra_pradesh_2014.csv")
data
```

```python
data[data["CONSTITUENCY"] == "Bhadrachalam (ST)"]
```

```python
const_list = list(data["CONSTITUENCY"].unique())
const_list
```

```python
if "nan" in const_list:
    print(True)
else:
    print(False)
```

```python
party_const = {}
for index,row in data.iterrows():
    if row["CONSTITUENCY"] in const_list:
        keys_list = list(party_const.keys())
        if row["PARTY"] not in keys_list:
            party_const[row["PARTY"]] = 1
        else:
            party_const[row["PARTY"]] = party_const[row["PARTY"]] + 1
party_const
```
