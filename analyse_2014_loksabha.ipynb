```python
constituency_list = list(data["CONSTITUENCY"].unique())
len(constituency_list)
```

```python
constituency_party = {}
win_party_constituency = {}
for const in constituency_list:
    only_const = data[data["CONSTITUENCY"] == const]
    party = only_const[only_const["% OF TOTAL VOTES"] == only_const["% OF TOTAL VOTES"].max()]
    ruling_party = list(party["PARTY"])
    ruling_party = ruling_party[0]
    constituency_party[const] = ruling_party
    keys_list = list(win_party_constituency.keys())
    if ruling_party not in keys_list:
        win_party_constituency[ruling_party] = 1
    else:
        win_party_constituency[ruling_party] = win_party_constituency[ruling_party] + 1
#win_party_constituency
```

```python
win_party_constituency
```

```python
parties_count = pd.DataFrame(columns=["count"])
parties_count["count"] = list(win_party_constituency.values())
parties_count.index = list(win_party_constituency.keys())
parties_count
```

```python
parties_count["count"].iplot(kind = "bar", xTitle = "Parties", yTitle = "No. of votes", title = "No of seats won by a party")
```
